```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity

# Create a LabelEncoder to encode movie titles, users, and genres
movie_encoder = LabelEncoder()
user_encoder = LabelEncoder()
genre_encoder = LabelEncoder()

df4['movie_idx'] = movie_encoder.fit_transform(df4['title'])
df4['user_idx'] = user_encoder.fit_transform(df4['userId'])
df4['genre_idx'] = genre_encoder.fit_transform(df4['stemmed_genres'])

# Create a user-movie rating matrix
num_users = len(user_encoder.classes_)
num_movies = len(movie_encoder.classes_)
user_movie_matrix = np.zeros((num_users, num_movies))

for row in df4.itertuples():
    user_movie_matrix[row.user_idx, row.movie_idx] = row.rating

# Calculate cosine similarity between movies (considering both ratings and genres)
movie_similarity = cosine_similarity(user_movie_matrix.T)  # You can combine this with genre similarity

# Define K for KNN
K = 20

# Recommend movies for a given movie index
def recommend_movies(movie_idx):
    similar_movies = np.argsort(movie_similarity[movie_idx])[-K-1:-1]  # Exclude the movie itself
    return similar_movies

# Example: Recommend movies for a specific movie index
movie_idx = movie_encoder.transform(['Iron Man (2008)'])[0]
recommended_movie_indices = recommend_movies(movie_idx)

recommended_movies = movie_encoder.inverse_transform(recommended_movie_indices)
print("Recommended movies:")
for movie_title in recommended_movies:
    print(movie_title)
```
